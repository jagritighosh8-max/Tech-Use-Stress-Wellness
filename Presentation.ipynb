```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
df = pd.read_csv(r"C:\Users\LENOVO\Downloads\archive (1)\Tech_Use_Stress_Wellness.csv")
print("--- Data Snapshot ---")
print(df.head())
```

```python
# Check data types and for any missing values
print(f"--- Data Info ---")
print(df.info())
```

```python
sns.scatterplot(data=df, 
                x='work_related_hours', 
                y='stress_level', 
                hue='mental_health_score', 
                palette='coolwarm', 
                alpha=0.7)
plt.title('Work Hours  Stress Level (colored by Mental Health Score)')
plt.xlabel('Hours Worked')
plt.ylabel('Stress Level')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='uses_wellness_apps', y='mental_health_score', palette='Set2')

plt.title('Impact of Wellness Apps on Mental Health Scores')
plt.xlabel('Uses Wellness Apps')
plt.ylabel('Mental Health Score')
plt.show()

avg_scores = df.groupby('uses_wellness_apps')['mental_health_score'].mean()
print("\n--- Average Mental Health Score by App Usage ---")
print(avg_scores)
```

```python
#Users who use wellness apps have a significantly higher median mental health score (~68) 
#compared to those who don't (~62). This validates the effectiveness of the product category.
```

```python
#Business Context: Who is most stressed? Is it a gender issue? A location issue? This helps in targeting marketing or support programs.
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='gender', y='stress_level', palette='pastel', errorbar=None)

plt.title('Average Stress Level by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Stress Level (0-10)')
plt.show()
```

```python
#Social Media is a Stressor: There is a near-perfect correlation (0.95) between social_media_hours and stress_level. Higher usage is strongly linked to higher stress.

#Wellness Apps Add Value: Individuals using wellness apps have, on average, a Mental Health Score that is 6 points higher than those who do not.

#Screen Time Trade-off: High screen time is negatively correlated with sleep duration (-0.75), which in turn hurts mental health.
```
